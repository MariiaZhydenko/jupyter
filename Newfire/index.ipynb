```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("survey_test.csv")
```

```python
row_yes_no = ['family_history','treatment', 'mental_health_consequence', 'phys_health_consequence', 
              'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview', 'obs_consequence', 
              'mental_vs_physical', 'remote_work', 'tech_company', 'benefits', 'care_options', 'wellness_program',
             'seek_help', 'anonymity']
row_leave = ['leave']
row_country = ['Country']
#replace 'usa', 'The USA' etc. => "USA"  
def parse_country(x):
    if 'usa' in x.lower() or x == "United States":
        return 'USA'
    return x

# Replace 'Yes', 'No' etc. => "1", "0"
# yes_no ={
#     'no': 0,
#     '0': 0,
#     'yes': 1,
#     'some of them': 2,
#     'maybe': 3,
#     "don't know": 4,
#     "Not sure": 5
# }
def parse_yes_no(x):
    if pd.isna(x):
        return None
    elif 'no' in x.lower():
        return 0
    elif 'yes' in x.lower():
        return 1
    elif 'some of them' in x.lower():
        return 2
    elif 'maybe' in x.lower():
        return 3
    elif "don't know" in x.lower():
        return 4
    elif "Not sure" in x.lower():
        return 5
    return None

# Replace leave
# Somewhat easy => 0
# Don't know => 1
# Somewhat difficult => 2
# Very difficult => 3
# Very easy => 4
def parse_leave(x):
    if pd.isna(x):
        return None
    elif 'somewhat easy' in x.lower():
        return 0
    elif "don't know" in x.lower():
        return 1
    elif 'somewhat difficult' in x.lower():
        return 2
    elif 'very difficult' in x.lower():
        return 3
    elif 'very easy' in x.lower():
        return 4
    return None

# Replace work_interfere
# Often => 0
# Don't know => 1
# Somewhat difficult => 2
# Very difficult => 3
# Very easy => 4
def parse_leave(x):
    if pd.isna(x):
        return None
    elif 'Often' in x.lower():
        return 0
    elif "don't know" in x.lower():
        return 1
    elif 'somewhat difficult' in x.lower():
        return 2
    elif 'very difficult' in x.lower():
        return 3
    elif 'very easy' in x.lower():
        return 4
    return None
```

```python
for d in df:
    if d in row_yes_no:
        df[d] = df[d].apply(parse_yes_no)
    elif d in row_leave:
        df[d] = df[d].apply(parse_leave)
    elif d in row_country:
        df[d] = df[d].apply(parse_country)
count = 1261
```

```python
del df['Timestamp']
del df['Age']
del df['Gender']
del df['state']
del df[u'comments']
del df['Country']
del df['work_interfere']
del df['no_employees']
del df['self_employed']
```

```python
data_value = df.apply(pd.value_counts).fillna(0)
```

```python
predict_column = []
for d in list(data_value.columns):
    data_value[d+"_percent"] = pd.Series(np.random.randn(5), index=data_value.index)
    for index, value in enumerate(data_value[d]):
        if value/count*100 > 70:
            predict_column.append(d)
        data_value.set_value(index, d+"_percent", str(value/count*100))
```
